# 중복재귀순열

    - 중복재귀순열은 인형뽑았다가 다시 인형을 넣는다.
    - 즉, 뽑았던 인형을 다시 뽑을 수 있음
    - 개수는 개수^개수

```python
arr = ['A', 'B', 'C', 'D']
sel = [0,0,0,0]
sel_list = []
def perm_rep(depth):
    if depth ==4:
        print([*sel])
        sel_list.append([*sel])
        return

    for i in range(4):
        sel[depth] = arr[i]
        perm_rep(depth+1)
    
perm_rep(0)
print(sel_list)
print(len(sel_list))
print(4*4*4*4)
```
