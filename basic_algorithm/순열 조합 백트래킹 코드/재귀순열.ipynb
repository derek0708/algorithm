# 재귀순열

 - 재귀순열은 중복이 안된다.
 - 재귀순열은 인형뽑기와 같다고 생각하면 된다 뽑는순간 out
 - 개수는 펙토리얼

```python
arr = ['A', 'B', 'C']
sel = [0, 0, 0]
check = [0, 0, 0]

def perm(depth):
    if depth == 3:
        print(sel)
        return
    
    for i in range(3):
        if not check[i]:
            check[i] = 1
            sel[depth] = arr[i]
            perm(depth+1)
            check[i] = 0

perm(0)
```

```python
arr = ['A', 'B', 'C', 'D', 'F']
sel = [0,0,0,0,0]
check = [0,0,0,0,0]
sel_list = []


def perm(depth):
    if depth == 5:
        print(sel)
        sel_list.append(sel)
        return
    for i in range(5):
        if check[i] == 0:
            check[i] = 1
            sel[depth] = arr[i]
            perm(depth + 1)
            check[i] = 0

perm(0)

print(sel_list)
print(len(sel_list))

```
